1. Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
class BankAccount:
    def __init__(self,account_number,account_holder_name,balance=0):
        self.account_number=account_number
        self.account_holder_name=account_holder_name
        self.balance=balance
        
    def deposit(self,amount):
        if amount>0:
            self.balance+=amount
            print(f"Deposited ${amount} into the account. New balance:${self.balance}")
        elif amount<=0:
            print("Invalid deposit amount.Please enter a positive amount.")
        
    def withdraw(self,amount):
        if amount>0 and amount<=self.balance:
            self.balance-=amount
            print(f"Withdrew ${amount} from the account. New balance: ${self.balance}")
        elif amount<=0:
            print("Invalid wihdrawal amount.Please enter a positive amount.")
        else:
            print("Insufficient funds.")
    def get_balance(self):
        return self.balance
    
    def __str__(self):
        return f"Account Number: {self.account_number}\nAccount Holder Name:{self.account_holder_name}\nBalance:${self.balance}"
        
# Example usage:
if __name__=="__main__":
    # Create a bank account
    
    account1=BankAccount("34433564","Baburao Apte")
    
    # Deposit and withdraw money
    account1.deposit(10000)
    account1.withdraw(5000)
    account1.withdraw(7000)  # Should print 'Insufficient funds.'
    
    # Get account information and balance
    
    print(account1)
```

2. Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

```python
class Employee:
    def __init__(self,employee_id,name,salary):
        self.employee_id=employee_id
        self.name=name
        self.salary=salary
    
    def calculate_yearly_bonus(self,bonus_percentage):
        if bonus_percentage>=0:
            bonus_amount=(bonus_percentage/100)* self.salary
            return bonus_amount
        else:
            return 0
        
    def display_employee_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"salary:{self.salary}")
        


# Example usage:
if __name__=="__main__":
    # Create an employee
    
    employee1=Employee(407,'Baburao Apte',40000)
    # Calculate and display yearly bonus
    bonus_percentage=15 # 10% bonus
    yearly_bonus=employee1.calculate_yearly_bonus(bonus_percentage)
    print(f"Yearly Bonus: ${yearly_bonus}")
    
    # Display employee details
    employee1.display_employee_details()
```

3. Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles={}
        
    def rent_vehicle(self,vehicle_id,vehicle_type):
        if vehicle_id not in self.available_vehicles:
            print(f"Sorry, vehicle {vehicle_id} is not available for rent.")
            
        else:
            self.available_vehicles.pop(vehicle_id)
            print(f"vehicle {vehicle_id} ({vehicle_type}) has been rented.")
            
    def return_vehicle(self,vehicle_id,vehicle_type):
        self.available_vehicles[vehicle_id]=vehicle_type
        print(f"Vehicle {vehicle_id} ({vehicle_type}) has benn returned.")
    
    
    def display_available_vehicles(self):
        if not self.available_vehicles:
            print("No vehicles are currently available.")
        else:
            print("Available Vehicles:")
            
            for vehicle_id, vehicle_type in self.available_vehicles.items():
                print(f"Vehicle ID: {vehicle_id}, Type: {vehicle_type}")
                


# Example Usage:
if __name__=="__main__":
    rental_system=VehicleRentalSystem()
    
    # Add some available vehicles to the system
    rental_system.available_vehicles={1:"Car", 2:"Scooter", 3: "Bicycle"}
    
    # Display available vehicles
    rental_system.display_available_vehicles()
    # Rent a vehicle
    rental_system.rent_vehicle(2,"Scooter")
    
    # Return a vehicle
    rental_system.return_vehicle(2,"Scooter")
    
    # Display available vehicles after returning
    
    rental_system.display_available_vehicles()
```

4. Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
class Book:
    def __init__(self,title, author, isbn):
        self.title=title
        self.author=author
        self.isbn=isbn
        self.is_available=True
    
    def borrow(self):
        if self.is_available:
            self.is_available=False
            return True
        else:
            return False
        
    def return_book(self):
        if not self.is_available:
            self.is_available=True
            
    def __str__(self):
        return f"{self.title} by {self.author}"
    
class Library:
    def __init__(self):
        self.books=[]
        
    def add_book(self,book):
        self.books.append(book)
    
    def borrow_book(self,isbn):
        for book in self.books:
            if book.isbn==isbn:
                if book.borrow():
                    print(f"You have successfully borrowed '{book.title}'.")
                else:
                    print(f"'{book.title}' is already borrowed.")
                return 
            print("Book with ISBN not found in the library.")
            
    def return_book(self,isbn):
        for book in self.books:
            if book.isbn==isbn:
                book.return_book()
                print(f"You have successfully returned '{book.title}'.")
                return
            print("Book with ISBN not found in the library.")
            
    def display_available_books(self):
        available_books=[book for book in self.books if book.is_available]
        if available_books:
            print("Available Books:")
            for book in available_books:
                print(book)
        else:
            print("No books are currently available in the library.")
            
# Example Usage:
if __name__=="__main__":
    library=Library()
    
    # Adding books to the library
    book1=Book("The Great Gatsby","F.Scott Fitzgerald","978-0743273565")
    book2=Book("To Kill a Mockingbird","Harper Lee","978-0061120084")
    book3=Book("1984","George Orwell","978-0451524935")
    
    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)
    
    # Display available books
    library.display_available_books()
    
    # Borrowing and returning books
    library.borrow_book("978-0743273565") 
    library.borrow_book("978-0743273565") # Already borrowed
    library.return_book("978-0743273565")
    library.return_book("978-0451524935")
    
    # Display available books after borrowing and returning
    
    library.display_available_books()   
```

5. Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id=product_id
        self.name=name
        self.price=price
        self.quantity=quantity
        
    def update_quantity(self, new_quantity):
        if new_quantity>=0:
            self.quantity=new_quantity
            print(f"Updated quantity of '{self.name}' to {new_quantity}.")
        else:
            print("Invalid quantity. Please enter a non-negative quantity.")
            
    def __str__(self):
        return f"Product ID: {self.product_id}\nName: {self.name}\nPrice: ${self.price}\nQuantity: {self.quantity}"
    
class Inventory:
    def __init__(self):
        self.products={}
    def add_product(self, product):
        if product.product_id not in self.products:
            self.products[product.product_id]=product
            print(f"'{product.name}' has been added to the inventory.")

        else:
            print(f"'{product_name}' already exists in the inventory.")

    def update_quantity(self, product_id, new_quantity):
        if product_id in self.products:
            self.products[product_id].update_quantity(new_quantity)
        else:
            print("Product not found in the inventory.")

    def display_available_products(self):
        if not self.products:
            print("No products are currently available in the inventory.")
        else:
            print("Available Products:")
            for product in self.products.values():
                print(product)
# Example usage:
if __name__=="__main__":
    inventory_system=Inventory()
    
    # Adding products to the inventory
    product1=Product("101","Laptop",800,10)
    product2=Product("102","Phone",400,20)
    product3=Product("103","Tablet",300,15)
    
    
    inventory_system.add_product(product1)
    inventory_system.add_product(product2)
    inventory_system.add_product(product3)
    
    # Display available products
    inventory_system.display_available_products()
    
    # Update product quantity
    inventory_system.update_quantity("101",5)
    
    # Display available products after updating quantity
    inventory_system.display_available_products()
```

6. Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length=length
        self.width=width
        self.height=height
        
    def calculate_area(self):
        pass
    
    def calculate_perimeter(self):
        pass
    
class Rectangle(Shape):
    def __init__(self, length, width):
        super().__init__(length, width)
        
    def calculate_area(self):
        return self.length*self.width
    
    def calculate_perimeter(self):
        return 2 * (self.length + self.width)
    
class Cube(Shape):
    def __init__(self, length):
        super().__init__(length, length, length)
        
    def calculate_area(self):
        return 6 * self.length **2
    def calculate_volume(self):
        return self.length**3
    
# Example usage:

if __name__=="__main__":
    
    rectangle=Rectangle(5,4)
    cube=Cube(3)
    
    print("Rectangle:")
    print(f"Area: {rectangle.calculate_area()}")
    print(f"Perimeter: {rectangle.calculate_perimeter()}")
    
    print("\nCube")
    print(f"Surface Area: {cube.calculate_area()}")
    print(f"Volume: {cube.calculate_volume()}")
```

7. Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
class Student:
    def __init__(self, student_id, name):
        self.student_id=student_id
        self.name=name
        self.grades=[]
        
    def add_grade(self, grade):
        self.grades.append(grade)
        
    def calculate_average_grade(self):
        if len(self.grades)==0:
            return 0
        total_grade=sum(self.grades)
        average_grade=total_grade/len(self.grades)
        return average_grade
    
    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        
        if len(self.grades)>0:
            print(f"Grades: {', '.join(map(str,self.grades))}")
            print(f"Average Grade: {self.calculate_average_grade(): .2f}")
            
        else:
            print("No grades available for this student.")
            
# Example usage:
if __name__=="__main__":
    
    student1=Student(1,"Baburao")
    student2=Student(2, "Shantabai")
    
    student1.add_grade(85)
    student1.add_grade(90)
    student1.add_grade(78)
    
    student2.add_grade(92)
    student2.add_grade(88)
    
    student1.display_student_details()
    print("\n")
    student2.display_student_details()
```

8. Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.

```python
class Email:
    def __init__(self, sender, recipient, subject):
        self.sender=sender
        self.recipient=recipient
        self.subject=subject
        self.message=None  # Placeholder for the email message
        
    def set_message(self, message):
        self.message=message
        
    def send_email(self):
        if self.message:
            print("Sending Email...")
            print(f"From: {self.sender}")
            print(f"To: {self.recipient}")
            print(f"Subject: {self.subject}")
            print("\nMessage:")
            print(self.message)
            print("\nEmail sent successfully!")
        else:
            print("Cannot send the email. Please set a message first.")
            
    def display_email_details(self):
        print("Email Details:")
        print(f"From: {self.sender}")
        print(f"To: {self.recipient}")
        print(f"Subject: {self.subject}")
        
        if self.message:
            print("Message:")
            print(self.message)
        else:
            print("No message set for this email.")

# Example usage:
if __name__=="__main__":
    email1= Email("sender@example.com", "recipient@example.com", "Hello, World!")
    email1.set_message("This is a test email.")
    
    email2=Email("Shalu@example.com", "Gangnam@example.com", "Meeting Agenda")
    email2.set_message("Here's the agenda for our upcoming meeting.")
    
    email1.send_email()
    print("\n")
    email2.display_email_details()       
```

9. Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username=username
        self.posts=[]
        
    def add_post(self, post_content):
        self.posts.append(post_content)
        print(f"New post added by {self.username}.")
        
    def display_posts(self):
        if not self.posts:
            print(f"{self.username} has no posts yet.")
            
        else:
            print(f"Posts by {self.username}:")
            for i, post in enumerate(self.posts, start=1):
                print(f"Match {i}: {post}")
    
    
    def search_posts_by_keyword(self, keyword):
        matching_posts=[post for post in self.posts if keyword.lower() in post.lower()]
        if not matching_posts:
            print(f"No posts by {self.username} contain the keyword '{keyword}'.")
        else:
            print(f"Posts by {self.username} containing the keyword '{keyword}':")
            for i, post in enumerate(matching_posts, start=1):
                print(f"Match {i}: {post}")
                
# Example usage:
if __name__=="__main__":
    profile1=SocialMediaProfile("alice_username")
    profile2=SocialMediaProfile("bob_username")
    
    profile1.add_post("Hello, this is my first post!")
    profile1.add_post("I'm enjoying my day.")
    
    profile2.add_post("Just had a great meal!")
    profile2.add_post("Heading out for a hike tomorrow.")
    
    # Display posts
    profile1.display_posts()
    print("\n")
    profile2.display_posts()
    
    # Search for posts by keyword
    profile1.search_posts_by_keyword("enjoying")
    profile2.search_posts_by_keyword("hike")
```

10. ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class ToDoList:
    def __init__(self):
        self.tasks={} # Dictionary to store tasks and their due dates
        
    def add_task(self, task, due_date):
        self.tasks[task]=due_date
        print(f"task '{task}' added with due date {due_date}.")
        
    def mark_task_completed(self, task):
        if task in self.tasks:
            del self.tasks[task]
            print(f"Task '{task}' marked as completed.")
        else:
            print(f"Task '{task}' not found in the ToDo list.")
            
    def display_pending_tasks(self):
        if not self.tasks:
            print("No pending tasks.")
        else:
            print("Pending Tasks:")
            for task, due_date in self.tasks.items():
                print(f"Task: {task}, Due Date: {due_date}")
                
# Example usage:
if __name__=="__main__":
    todo_list=ToDoList()
    
    todo_list.add_task("Buy groceries", "2023-10-15")
    todo_list.add_task("Finish project report", "2023-10-20")
    todo_list.add_task("Call Mike", "2023-10-25")
    
    todo_list.display_pending_tasks()
    print("\n")
    
    todo_list.mark_task_completed("Buy groceries")
    todo_list.display_pending_tasks()
```
